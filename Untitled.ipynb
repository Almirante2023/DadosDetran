```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl


arquivos = [
    'datatran2007.csv', 'datatran2008.csv', 'datatran2009.csv',
    'datatran2010.csv', 'datatran2011.csv', 'datatran2012.csv',
    'datatran2013.csv', 'datatran2014.csv', 'datatran2015.csv',
    'datatran2016.csv', 'datatran2017.csv', 'datatran2018.csv',
    'datatran2020.csv'
]


dataframes = []


for arquivo in arquivos:
    df_temp = pd.read_csv(arquivo, sep=';', encoding='latin1')
    dataframes.append(df_temp)


df_completo = pd.concat(dataframes)


display(df_completo)
```

```python
tipo_acidente_contagem = df_completo['tipo_acidente'].value_counts()

display(tipo_acidente_contagem)

tipo_acidente_contagem = tipo_acidente_contagem.sort_values(ascending=False)
```

```python


# Supondo que 'df_completo' seja o DataFrame com os dados e 'tipo_acidente' seja a coluna dos tipos de acidente
tipo_acidente_contagem = df_completo['tipo_acidente'].value_counts()

# Ordenar do maior para o menor e selecionar os 20 maiores
tipo_acidente_contagem = tipo_acidente_contagem.sort_values(ascending=False).head(10)

# Plotar um gráfico de barras com Pandas para os 20 maiores tipos de acidente
tipo_acidente_contagem.plot(kind='bar', figsize=(10, 6), color='skyblue')

plt.xlabel('Tipos de Acidente')  # Definindo o rótulo do eixo x
plt.ylabel('Quantidade de Ocorrências')  # Definindo o rótulo do eixo y
plt.title('10 Maiores Tipos de Acidente por Ocorrências')  # Definindo o título do gráfico
plt.xticks(rotation=45)  # Rotacionando os rótulos do eixo x para melhor visualização
plt.tight_layout()  # Ajustando o layout para evitar cortes
plt.show()
```

```python
dia_dos_acidentes =  df_completo['dia_semana'].value_counts()

dia_dos_acidentes = dia_dos_acidentes.sort_values(ascending=False)

display(dia_dos_acidentes)
```
